```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Estilo visual
sns.set(style="whitegrid", palette="pastel")

# Diccionario de preguntas
preguntas_dict = {
    "Horas de sueo (semana)": "Cuntas horas duermes en un da tpico durante la semana?",
    "Horas de clases (semana)": "Cuntas horas tienes clases un da tpico en la semana?",
    "Horas de estudio/tareas (semana)": "Cuntas horas estudias o haces tareas en un da tpico en la semana?",
    "Horas extracurriculares (semana)": "Cuntas horas haces actividades extracurriculares en un da tpico?",
    "Horas pantallas recreativas (semana)": "Cuntas horas usas pantallas recreativamente en un da tpico en la semana?",
    "Horas descanso sin pantallas (semana)": "Cuntas horas descansas sin pantallas en un da tpico en la semana?",
    "Horas de sueo (fin de semana)": "Cuntas horas duermes durante un da tpico el fin de semana?",
    "Horas de estudio/tareas (fin de semana)": "Cuntas horas estudias o haces tareas durante un da tpico el fin de semana?",
    "Horas pantallas recreativas (fin de semana)": "Cuntas horas usas pantallas recreativamente durante un da tpico en el finde?",
    "Horas con familia o amistades (fin de semana)": "Cuntas horas compartes con familia o amistades durante un da tpico el finde?",
    "Horas autocuidado (fin de semana)": "Cuntas horas dedicas a actividades fsicas o de autocuidado durante un da tpico el finde?"
}

# Orden lgico de cursos universitarios
orden = [
    "Primer ao de universidad", "Segundoao de universidad",
    "Tercer ao de universidad", "Cuartoao de universidad",
    "Quinto ao de universidad"
]

# Leer y reorganizar base universitaria
data_estudiantes = pd.read_csv('data_estudiantes_ingenieria.csv')
df_universitarios = data_estudiantes.copy()
df_universitarios = df_universitarios.melt(
    id_vars=["Curso o ao actual"],
    value_vars=list(preguntas_dict.keys()),
    var_name="Pregunta",
    value_name="Horas"
)

# Filtrar solo los cursos universitarios definidos
df_universitarios = df_universitarios[
    df_universitarios["Curso o ao actual"].isin(orden)
]

# Asegurar orden en el eje X
df_universitarios["Curso o ao actual"] = pd.Categorical(
    df_universitarios["Curso o ao actual"], categories=orden, ordered=True
)

# Grfico por cada pregunta
for columna, pregunta in preguntas_dict.items():
    plt.figure(figsize=(13, 6))

    # Subset para grfica
    subset = df_universitarios[df_universitarios["Pregunta"] == columna]

    # Dispersin por curso
    sns.stripplot(
        data=subset,
        x="Curso o ao actual",
        y="Horas",
        hue="Curso o ao actual",
        jitter=0.25,
        dodge=False,
        alpha=0.6,
        size=7,
        palette="Set2",
        legend=False
    )

    # Promedio por curso
    sns.pointplot(
        data=subset,
        x="Curso o ao actual",
        y="Horas",
        estimator=np.median,
        color="black",
        markers="D",
        linestyles="--",
        scale=1.5,
        errorbar=None
    )

    # Ttulos y estilo
    plt.title(pregunta, fontsize=16, fontweight="bold")
    plt.xlabel("Curso o ao actual", fontsize=12)
    plt.ylabel("Horas", fontsize=12)
    plt.xticks(rotation=35)
    plt.grid(True, linestyle="--", alpha=0.4)
    plt.tight_layout()
    plt.show()
```
